```python
dados = [

    # VT = salto; UB = barras; BB = trave; FX = solo
    # Quando a ginasta aparece em mais de um aparelho no mesmo evento,
    # preencho na mesma linha para permitir somatória parcial.

    # MUNDIAL 2025 - FINAL AA

    ["Mundial 2025", 2025, "mundial", "AA", "Angelina Melnikova", "AIN/Rússia", None, None, None, None, 55.066],
    ["Mundial 2025", 2025, "mundial", "AA", "Leanne Wong", "Estados Unidos", None, None, None, None, 54.966],
    ["Mundial 2025", 2025, "mundial", "AA", "Zhang Qingying", "China", None, None, None, None, 54.633],
    ["Mundial 2025", 2025, "mundial", "AA", "Kaylia Nemour", "Argélia", None, None, None, None, 54.564],
    ["Mundial 2025", 2025, "mundial", "AA", "Asia D'Amato", "Itália", None, None, None, None, 53.532],
    ["Mundial 2025", 2025, "mundial", "AA", "Rina Kishi", "Japão", None, None, None, None, 53.232],
    ["Mundial 2025", 2025, "mundial", "AA", "Aiko Sugihara", "Japão", None, None, None, None, 53.132],
    ["Mundial 2025", 2025, "mundial", "AA", "Abigail Martin", "Reino Unido", None, None, None, None, 52.998],


    # PAN-AMERICANO 2025 - AA

    ["Pan-Americano 2025", 2025, "continental", "AA", "Jayla Hang", "Estados Unidos", None, None, None, None, 55.300],
    ["Pan-Americano 2025", 2025, "continental", "AA", "Lia Monica Fontaine", "Canadá", None, None, None, None, 53.966],
    ["Pan-Americano 2025", 2025, "continental", "AA", "Hezly Rivera", "Estados Unidos", None, None, None, None, 52.667],
    ["Pan-Americano 2025", 2025, "continental", "AA", "Dulcy Caylor", "Estados Unidos", None, None, None, None, 52.600],
    ["Pan-Americano 2025", 2025, "continental", "AA", "Gabrielle Black", "Canadá", None, None, None, None, 51.433],
    ["Pan-Americano 2025", 2025, "continental", "AA", "Alyssa Guerrier-Calixte", "Canadá", None, None, None, None, 50.100],
    ["Pan-Americano 2025", 2025, "continental", "AA", "Emilia Acosta", "Argentina", None, None, None, None, 49.966],
    ["Pan-Americano 2025", 2025, "continental", "AA", "Isabella Ajalla", "Argentina", None, None, None, None, 49.667],
    ["Pan-Americano 2025", 2025, "continental", "AA", "Thais Fidélis", "Brasil", None, None, None, None, 49.467],
    ["Pan-Americano 2025", 2025, "continental", "AA", "Gabriela Barbosa", "Brasil", None, None, None, None, 49.300],
    ["Pan-Americano 2025", 2025, "continental", "AA", "Karla Navas", "Panamá", None, None, None, None, 48.500],
    ["Pan-Americano 2025", 2025, "continental", "AA", "Ginna Escobar", "Colombia", None, None, None, None, 48.467],

    # ASIÁTICO 2025 - AA

    ["Asiático 2025", 2025, "continental", "AA", "Aiko Sugihara", "Japão", None, None, None, None, 54.765],
    ["Asiático 2025", 2025, "continental", "AA", "Haruka Nakamura", "Japão", None, None, None, None, 54.065],
    ["Asiático 2025", 2025, "continental", "AA", "Qin Xinyi", "China", None, None, None, None, 53.599],
    ["Asiático 2025", 2025, "continental", "AA", "Tonya Paulsson", "Taiwan", None, None, None, None, 52.499],
    ["Asiático 2025", 2025, "continental", "AA", "Zhang Xinyi", "China", None, None, None, None, 51.999],
    ["Asiático 2025", 2025, "continental", "AA", "Park Na-young", "Coreia do Sul", None, None, None, None, 51.166],
    ["Asiático 2025", 2025, "continental", "AA", "Eom Do-hyun", "Coreia do Sul", None, None, None, None, 50.865],
    ["Asiático 2025", 2025, "continental", "AA", "Yeap Kang Xian", "Malásia", None, None, None, None, 49.433],


    # COPAS / CHALLENGE CUPS 2025 - FINAIS POR APARELHO
    # COTTBUS WORLD CUP 2025

    ["Cottbus World Cup 2025", 2025, "world_cup", "APARELHO", "Teja Belak", "Eslovênia", 13.299, None, None, None, None],
    ["Cottbus World Cup 2025", 2025, "world_cup", "APARELHO", "Tjasa Kysselef", "Eslovênia", 13.266, None, None, None, None],
    ["Cottbus World Cup 2025", 2025, "world_cup", "APARELHO", "Oksana Chusovitina", "Uzbequistão", 13.249, None, None, None, None],
    ["Cottbus World Cup 2025", 2025, "world_cup", "APARELHO", "Christine Kubon", "Noruega", 13.033, None, None, None, None],
    ["Cottbus World Cup 2025", 2025, "world_cup", "APARELHO", "Bilge Tarhan", "Turquia", 12.633, None, None, 11.233, None],
    ["Cottbus World Cup 2025", 2025, "world_cup", "APARELHO", "Keisha Lockert", "Noruega", 12.516, None, None, 11.100, None],
    ["Cottbus World Cup 2025", 2025, "world_cup", "APARELHO", "Katrina Jurevica", "Letônia", 12.433, None, None, None, None],
    ["Cottbus World Cup 2025", 2025, "world_cup", "APARELHO", "Daniz Aliyeva", "Azerbaijão", 12.383, None, None, None, None],

    ["Cottbus World Cup 2025", 2025, "world_cup", "APARELHO", "Zhang Kexin", "China", None, 13.900, 14.266, 12.633, None],
    ["Cottbus World Cup 2025", 2025, "world_cup", "APARELHO", "Charlotte Booth", "Reino Unido", None, 13.300, None, 10.000, None],
    ["Cottbus World Cup 2025", 2025, "world_cup", "APARELHO", "Yamada Chiharu", "Japão", None, 13.033, 13.466, None, None],
    ["Cottbus World Cup 2025", 2025, "world_cup", "APARELHO", "Emily Roper", "Reino Unido", None, 12.533, None, None, None],
    ["Cottbus World Cup 2025", 2025, "world_cup", "APARELHO", "Zhang Yihan", "China", None, 12.533, None, 13.433, None],
    ["Cottbus World Cup 2025", 2025, "world_cup", "APARELHO", "Bianca Frysak", "Áustria", None, 12.300, 11.766, None, None],
    ["Cottbus World Cup 2025", 2025, "world_cup", "APARELHO", "Lihie Raz", "Israel", None, 11.733, None, 12.933, None],
    ["Cottbus World Cup 2025", 2025, "world_cup", "APARELHO", "Silja Stöhr", "Alemanha", None, 11.300, None, None, None],

    ["Cottbus World Cup 2025", 2025, "world_cup", "APARELHO", "Zhou Yaqin", "China", None, None, 14.766, None, None],
    ["Cottbus World Cup 2025", 2025, "world_cup", "APARELHO", "Lai Pin-Ju", "Taiwan", None, None, 13.066, None, None],
    ["Cottbus World Cup 2025", 2025, "world_cup", "APARELHO", "Diana Lobok", "Ucrânia", None, None, 12.433, 11.066, None],
    ["Cottbus World Cup 2025", 2025, "world_cup", "APARELHO", "Christina Zwicker", "Croácia", None, None, 11.866, None, None],
    ["Cottbus World Cup 2025", 2025, "world_cup", "APARELHO", "Lin Yi-Chen", "Taiwan", None, None, 11.800, None, None],
    ["Cottbus World Cup 2025", 2025, "world_cup", "APARELHO", "Maria Drobniak", "Polônia", None, None, None, 12.300, None],

    # BAKU WORLD CUP 2025
    ["Baku World Cup 2025", 2025, "world_cup", "APARELHO", "Oksana Chusovitina", "Uzbequistão", 13.516, None, None, None, None],
    ["Baku World Cup 2025", 2025, "world_cup", "APARELHO", "Teja Belak", "Eslovênia", 13.366, None, None, None, None],
    ["Baku World Cup 2025", 2025, "world_cup", "APARELHO", "Tjasa Kysselef", "Eslovênia", 13.283, None, None, None, None],
    ["Baku World Cup 2025", 2025, "world_cup", "APARELHO", "Greta Mayer", "Hungria", 13.266, None, 13.133, None, None],
    ["Baku World Cup 2025", 2025, "world_cup", "APARELHO", "Protistha Samanta", "Índia", 12.766, None, None, None, None],
    ["Baku World Cup 2025", 2025, "world_cup", "APARELHO", "Bengisu Yildiz", "Turquia", 12.666, None, 12.200, 12.266, None],
    ["Baku World Cup 2025", 2025, "world_cup", "APARELHO", "Daniz Aliyeva", "Azerbaijão", 12.483, None, None, None, None],
    ["Baku World Cup 2025", 2025, "world_cup", "APARELHO", "Nishka Agarwal", "Índia", 12.149, None, None, None, None],

    ["Baku World Cup 2025", 2025, "world_cup", "APARELHO", "Alena Tsitavets", "Belarus", None, 13.400, None, None, None],
    ["Baku World Cup 2025", 2025, "world_cup", "APARELHO", "Lucija Hribar", "Eslovênia", None, 13.233, None, None, None],
    ["Baku World Cup 2025", 2025, "world_cup", "APARELHO", "Tokutsugi Ayu", "Japão", None, 13.200, None, 12.900, None],
    ["Baku World Cup 2025", 2025, "world_cup", "APARELHO", "Sara Sulekic", "Croácia", None, 13.033, None, None, None],
    ["Baku World Cup 2025", 2025, "world_cup", "APARELHO", "Nazli Savranbasi", "Turquia", None, 13.000, 10.700, None, None],
    ["Baku World Cup 2025", 2025, "world_cup", "APARELHO", "Chu Yiming", "China", None, 12.566, 13.266, 12.766, None],
    ["Baku World Cup 2025", 2025, "world_cup", "APARELHO", "Nakamura Haruka", "Japão", None, 12.500, 13.433, None, None],
    ["Baku World Cup 2025", 2025, "world_cup", "APARELHO", "Sofia Mester", "Croácia", None, 12.433, None, None, None],

    ["Baku World Cup 2025", 2025, "world_cup", "APARELHO", "Kishi Rina", "Japão", None, None, 12.866, 13.066, None],
    ["Baku World Cup 2025", 2025, "world_cup", "APARELHO", "Angel Wong Hiu Ying", "Hong Kong", None, None, 11.433, None, None],
    ["Baku World Cup 2025", 2025, "world_cup", "APARELHO", "Ritu Das", "Índia", None, None, 10.200, None, None],
    ["Baku World Cup 2025", 2025, "world_cup", "APARELHO", "Arina Lebedeva", "Cazaquistão", None, None, None, 12.633, None],
    ["Baku World Cup 2025", 2025, "world_cup", "APARELHO", "Kira Maharevich", "Belarus", None, None, None, 12.233, None],
    ["Baku World Cup 2025", 2025, "world_cup", "APARELHO", "Natalia Jacoby", "Sérvia", None, None, None, 12.166, None],
    ["Baku World Cup 2025", 2025, "world_cup", "APARELHO", "Antea Sikic Kaucic", "Croácia", None, None, None, 12.100, None],

    # ANTALYA WORLD CUP 2025
    ["Antalya World Cup 2025", 2025, "world_cup", "APARELHO", "Jayla Hang", "Estados Unidos", 13.666, 12.733, 12.300, 12.966, None],
    ["Antalya World Cup 2025", 2025, "world_cup", "APARELHO", "Claire Pease", "Estados Unidos", 13.566, 14.233, 13.266, 13.266, None],
    ["Antalya World Cup 2025", 2025, "world_cup", "APARELHO", "Pranati Nayak", "Índia", 13.417, None, None, None, None],
    ["Antalya World Cup 2025", 2025, "world_cup", "APARELHO", "Tjasa Kysselef", "Eslovênia", 13.367, None, None, None, None],
    ["Antalya World Cup 2025", 2025, "world_cup", "APARELHO", "Oksana Chusovitina", "Uzbequistão", 13.366, None, None, None, None],
    ["Antalya World Cup 2025", 2025, "world_cup", "APARELHO", "Greta Mayer", "Hungria", 13.167, None, None, None, None],
    ["Antalya World Cup 2025", 2025, "world_cup", "APARELHO", "Julia Weissenhofer", "Liechtenstein", 13.000, None, None, None, None],
    ["Antalya World Cup 2025", 2025, "world_cup", "APARELHO", "Bengisu Yildiz", "Turquia", 12.367, None, 12.300, None, None],

    ["Antalya World Cup 2025", 2025, "world_cup", "APARELHO", "Yang Fanyuwei", "China", None, 14.800, None, None, None],
    ["Antalya World Cup 2025", 2025, "world_cup", "APARELHO", "Tian Zhuofan", "China", None, 13.866, 12.900, None, None],
    ["Antalya World Cup 2025", 2025, "world_cup", "APARELHO", "Lucija Hribar", "Eslovênia", None, 12.933, None, None, None],
    ["Antalya World Cup 2025", 2025, "world_cup", "APARELHO", "Nazli Savranbasi", "Turquia", None, 12.900, None, None, None],
    ["Antalya World Cup 2025", 2025, "world_cup", "APARELHO", "Emese Szij", "Hungria", None, 12.166, None, None, None],
    ["Antalya World Cup 2025", 2025, "world_cup", "APARELHO", "Paulina Guerra", "México", None, 11.366, None, None, None],

    ["Antalya World Cup 2025", 2025, "world_cup", "APARELHO", "Okamura Mana", "Japão", None, None, 13.133, 13.066, None],
    ["Antalya World Cup 2025", 2025, "world_cup", "APARELHO", "Sugihara Aiko", "Japão", None, None, 12.500, 13.633, None],
    ["Antalya World Cup 2025", 2025, "world_cup", "APARELHO", "Michelle Pineda", "México", None, None, 12.433, 13.266, None],
    ["Antalya World Cup 2025", 2025, "world_cup", "APARELHO", "Lai Pin-Ju", "Taiwan", None, None, 10.600, None, None],
    ["Antalya World Cup 2025", 2025, "world_cup", "APARELHO", "Qin Xinyi", "China", None, None, None, 13.000, None],
    ["Antalya World Cup 2025", 2025, "world_cup", "APARELHO", "Sara Peter", "Hungria", None, None, None, 12.700, None],
    ["Antalya World Cup 2025", 2025, "world_cup", "APARELHO", "Keisha Lockert", "Noruega", None, None, None, 12.233, None],

    # OSIJEK WORLD CUP 2025
    ["Osijek World Cup 2025", 2025, "world_cup", "APARELHO", "Valentina Georgieva", "Bulgária", 14.149, None, None, None, None],
    ["Osijek World Cup 2025", 2025, "world_cup", "APARELHO", "Yu Linmin", "China", 14.066, None, None, None, None],
    ["Osijek World Cup 2025", 2025, "world_cup", "APARELHO", "Jayla Hang", "Estados Unidos", 13.766, 11.666, 13.766, 13.466, None],
    ["Osijek World Cup 2025", 2025, "world_cup", "APARELHO", "Shallon Olsen", "Canadá", 13.566, None, None, None, None],
    ["Osijek World Cup 2025", 2025, "world_cup", "APARELHO", "Selina Kickinger", "Áustria", 13.216, None, 13.066, 12.700, None],
    ["Osijek World Cup 2025", 2025, "world_cup", "APARELHO", "Tijana Korent", "Croácia", 13.083, None, None, None, None],

    ["Osijek World Cup 2025", 2025, "world_cup", "APARELHO", "Yang Fanyuwei", "China", None, 13.766, None, None, None],
    ["Osijek World Cup 2025", 2025, "world_cup", "APARELHO", "Maisa Kuusikko", "Finlândia", None, 13.566, 12.300, None, None],
    ["Osijek World Cup 2025", 2025, "world_cup", "APARELHO", "Kate McDonald", "Austrália", None, 13.400, 12.133, None, None],
    ["Osijek World Cup 2025", 2025, "world_cup", "APARELHO", "Sanna Veerman", "Holanda", None, 13.233, None, None, None],
    ["Osijek World Cup 2025", 2025, "world_cup", "APARELHO", "Sara Sulekic", "Croácia", None, 13.100, None, None, None],
    ["Osijek World Cup 2025", 2025, "world_cup", "APARELHO", "Rachel Yeoh Li Wen", "Malásia", None, 13.033, None, None, None],
    ["Osijek World Cup 2025", 2025, "world_cup", "APARELHO", "Lilla Makai", "Hungria", None, 11.766, None, None, None],

    ["Osijek World Cup 2025", 2025, "world_cup", "APARELHO", "Eythora Thorsdottir", "Holanda", None, None, 13.900, 12.200, None],
    ["Osijek World Cup 2025", 2025, "world_cup", "APARELHO", "Jennifer Williams", "Suécia", None, None, 13.300, None, None],
    ["Osijek World Cup 2025", 2025, "world_cup", "APARELHO", "Cadence Teh Zi Qi", "Malásia", None, None, 12.533, None, None],
    ["Osijek World Cup 2025", 2025, "world_cup", "APARELHO", "Naomi Visser", "Holanda", None, None, 10.800, None, None],
    ["Osijek World Cup 2025", 2025, "world_cup", "APARELHO", "Yali Shoshani", "Israel", None, None, None, 13.300, None],
    ["Osijek World Cup 2025", 2025, "world_cup", "APARELHO", "Jemima Taylor", "Reino Unido", None, None, None, 12.600, None],
    ["Osijek World Cup 2025", 2025, "world_cup", "APARELHO", "Nazli Savranbasi", "Turquia", None, None, None, 12.433, None],
    ["Osijek World Cup 2025", 2025, "world_cup", "APARELHO", "Charlize Mörz", "Áustria", None, None, None, 11.766, None],
    ["Osijek World Cup 2025", 2025, "world_cup", "APARELHO", "Sona Artamonova", "Tchéquia", None, None, None, 11.633, None],

    # DOHA WORLD CUP 2025
    ["Doha World Cup 2025", 2025, "world_cup", "APARELHO", "Teja Belak", "Eslovênia", 13.316, None, None, None, None],
    ["Doha World Cup 2025", 2025, "world_cup", "APARELHO", "Tjasa Kysselef", "Eslovênia", 13.216, None, None, None, None],
    ["Doha World Cup 2025", 2025, "world_cup", "APARELHO", "Greta Mayer", "Hungria", 13.166, None, 13.133, 12.166, None],
    ["Doha World Cup 2025", 2025, "world_cup", "APARELHO", "Sara Peter", "Hungria", 13.166, None, None, 12.833, None],
    ["Doha World Cup 2025", 2025, "world_cup", "APARELHO", "Tijana Korent", "Croácia", 13.100, None, None, None, None],
    ["Doha World Cup 2025", 2025, "world_cup", "APARELHO", "Oksana Chusovitina", "Uzbequistão", 12.999, None, None, None, None],
    ["Doha World Cup 2025", 2025, "world_cup", "APARELHO", "Madeleine Marshall", "Nova Zelândia", 12.716, None, None, None, None],
    ["Doha World Cup 2025", 2025, "world_cup", "APARELHO", "Medina Kapbassova", "Cazaquistão", 12.566, None, None, 11.966, None],

    ["Doha World Cup 2025", 2025, "world_cup", "APARELHO", "Kate McDonald", "Austrália", None, 13.100, 10.933, 12.666, None],
    ["Doha World Cup 2025", 2025, "world_cup", "APARELHO", "Lucija Hribar", "Eslovênia", None, 13.033, None, None, None],
    ["Doha World Cup 2025", 2025, "world_cup", "APARELHO", "Audrey Rousseau", "Canadá", None, 12.700, 12.266, 12.833, None],
    ["Doha World Cup 2025", 2025, "world_cup", "APARELHO", "Diana Lobok", "Ucrânia", None, 11.566, 10.166, None, None],
    ["Doha World Cup 2025", 2025, "world_cup", "APARELHO", "Joanne Chen Hoi Yuen", "Hong Kong", None, 10.566, None, None, None],
    ["Doha World Cup 2025", 2025, "world_cup", "APARELHO", "Vita Prijanovic", "Eslovênia", None, 10.466, 11.566, None, None],
    ["Doha World Cup 2025", 2025, "world_cup", "APARELHO", "Jun McDonald", "Nova Zelândia", None, 10.200, None, None, None],
    ["Doha World Cup 2025", 2025, "world_cup", "APARELHO", "Polina Diachenko", "Ucrânia", None, 10.000, None, None, None],

    ["Doha World Cup 2025", 2025, "world_cup", "APARELHO", "Tina Zelcic", "Croácia", None, None, 12.766, None, None],
    ["Doha World Cup 2025", 2025, "world_cup", "APARELHO", "Christina Zwicker", "Croácia", None, None, 12.266, None, None],
    ["Doha World Cup 2025", 2025, "world_cup", "APARELHO", "Sabrina Voinea", "Romênia", None, None, 11.600, 13.600, None],
    ["Doha World Cup 2025", 2025, "world_cup", "APARELHO", "Odinakhon Robidjonova", "Uzbequistão", None, None, None, 12.266, None],
    ["Doha World Cup 2025", 2025, "world_cup", "APARELHO", "Amber Ward Wen Si", "Hong Kong", None, None, None, 11.833, None],

    # SZOMBATHELY CHALLENGE CUP 2025
    ["Szombathely Challenge Cup 2025", 2025, "world_cup", "APARELHO", "Charlize Mörz", "Áustria", 13.625, None, None, None, None],
    ["Szombathely Challenge Cup 2025", 2025, "world_cup", "APARELHO", "Greta Mayer", "Hungria", 13.400, None, 13.100, None, None],
    ["Szombathely Challenge Cup 2025", 2025, "world_cup", "APARELHO", "Tijana Korent", "Croácia", 13.050, None, None, None, None],
    ["Szombathely Challenge Cup 2025", 2025, "world_cup", "APARELHO", "Bilge Tarhan", "Turquia", 12.800, None, None, None, None],
    ["Szombathely Challenge Cup 2025", 2025, "world_cup", "APARELHO", "Valentina Georgieva", "Bulgária", 12.650, None, None, None, None],
    ["Szombathely Challenge Cup 2025", 2025, "world_cup", "APARELHO", "Djenna Laroui", "França", 12.300, None, None, None, None],
    ["Szombathely Challenge Cup 2025", 2025, "world_cup", "APARELHO", "Marina Escudero", "Espanha", 11.850, None, None, None, None],
    ["Szombathely Challenge Cup 2025", 2025, "world_cup", "APARELHO", "Tjasa Kysselef", "Eslovênia", 11.800, None, None, None, None],

    ["Szombathely Challenge Cup 2025", 2025, "world_cup", "APARELHO", "Zoja Szekely", "Hungria", None, 13.650, None, None, None],
    ["Szombathely Challenge Cup 2025", 2025, "world_cup", "APARELHO", "Romane Hamelin", "França", None, 13.250, None, None, None],
    ["Szombathely Challenge Cup 2025", 2025, "world_cup", "APARELHO", "Ana Luiza Lima", "Brasil", None, 13.050, None, None, None],
    ["Szombathely Challenge Cup 2025", 2025, "world_cup", "APARELHO", "Shanna-Kae Grant", "Reino Unido", None, 12.700, None, 12.050, None],
    ["Szombathely Challenge Cup 2025", 2025, "world_cup", "APARELHO", "Astria Nélo", "França", None, 12.600, None, None, None],
    ["Szombathely Challenge Cup 2025", 2025, "world_cup", "APARELHO", "Zsofia Kovacs", "Hungria", None, 12.050, None, None, None],
    ["Szombathely Challenge Cup 2025", 2025, "world_cup", "APARELHO", "Lucy Stewart", "Austrália", None, 11.350, None, 11.450, None],
    ["Szombathely Challenge Cup 2025", 2025, "world_cup", "APARELHO", "Laura Casabuena", "Espanha", None, 10.650, None, None, None],

    ["Szombathely Challenge Cup 2025", 2025, "world_cup", "APARELHO", "Flavia Saraiva", "Brasil", None, None, 13.800, None, None],
    ["Szombathely Challenge Cup 2025", 2025, "world_cup", "APARELHO", "Alba Petisco", "Espanha", None, None, 13.250, 11.400, None],
    ["Szombathely Challenge Cup 2025", 2025, "world_cup", "APARELHO", "Isabella Ajalla", "Argentina", None, None, 12.750, None, None],
    ["Szombathely Challenge Cup 2025", 2025, "world_cup", "APARELHO", "Diana Lobok", "Ucrânia", None, None, 12.750, None, None],
    ["Szombathely Challenge Cup 2025", 2025, "world_cup", "APARELHO", "Julia Soares", "Brasil", None, None, 12.700, 12.550, None],
    ["Szombathely Challenge Cup 2025", 2025, "world_cup", "APARELHO", "Alia Leat", "Reino Unido", None, None, 12.400, None, None],
    ["Szombathely Challenge Cup 2025", 2025, "world_cup", "APARELHO", "Philippa Busuttil", "Malta", None, None, 11.450, None, None],
    ["Szombathely Challenge Cup 2025", 2025, "world_cup", "APARELHO", "Denisa Golgota", "Romênia", None, None, None, 12.750, None],
    ["Szombathely Challenge Cup 2025", 2025, "world_cup", "APARELHO", "Julia Coutinho", "Brasil", None, None, None, 12.250, None],
    ["Szombathely Challenge Cup 2025", 2025, "world_cup", "APARELHO", "Lucia Gonzalez", "Argentina", None, None, None, 11.850, None],
    ["Szombathely Challenge Cup 2025", 2025, "world_cup", "APARELHO", "Emilia Acosta", "Argentina", None, None, None, 11.800, None],

    # JESOLO 2026 - SENIOR AA

    ["City of Jesolo Trophy 2026", 2026, "internacional", "AA", "Elena Colas", "França", 13.333, 14.333, 13.733, 12.633, 54.032],
    ["City of Jesolo Trophy 2026", 2026, "internacional", "AA", "Charleigh Bullock", "Estados Unidos", 13.433, 14.133, 12.533, 12.867, 52.966],
    ["City of Jesolo Trophy 2026", 2026, "internacional", "AA", "Li Rongjinyi", "China", 13.133, 13.733, 13.467, 12.567, 52.900],
    ["City of Jesolo Trophy 2026", 2026, "internacional", "AA", "Simone Rose", "Estados Unidos", 13.733, 12.867, 13.267, 13.000, 52.867],
    ["City of Jesolo Trophy 2026", 2026, "internacional", "AA", "Reese Esponda", "Estados Unidos", 14.000, 12.833, 12.133, 13.700, 52.666],
    ["City of Jesolo Trophy 2026", 2026, "internacional", "AA", "Caroline Moreau", "Estados Unidos", 12.967, 13.433, 13.567, 12.667, 52.634],

    # COPAS DO MUNDO 2026 - FINAIS POR APARELHO

    # COTTBUS WORLD CUP 2026

    ["Cottbus World Cup 2026", 2026, "world_cup", "APARELHO", "Anna Kalmykova", "Rússia", 13.816, None, None, 13.333, None],
    ["Cottbus World Cup 2026", 2026, "world_cup", "APARELHO", "Karina Schönmaier", "Alemanha", 13.900, None, None, None, None],
    ["Cottbus World Cup 2026", 2026, "world_cup", "APARELHO", "Teja Belak", "Eslovênia", 13.516, None, None, None, None],
    ["Cottbus World Cup 2026", 2026, "world_cup", "APARELHO", "Ushioku Kohane", "Japão", 13.333, None, None, None, None],
    ["Cottbus World Cup 2026", 2026, "world_cup", "APARELHO", "Lyudmila Roshchina", "Rússia", 13.316, None, None, None, None],
    ["Cottbus World Cup 2026", 2026, "world_cup", "APARELHO", "Emma Fioravanti", "Itália", 13.233, None, 13.233, 12.600, None],
    ["Cottbus World Cup 2026", 2026, "world_cup", "APARELHO", "Oksana Chusovitina", "Uzbequistão", 12.999, None, None, None, None],
    ["Cottbus World Cup 2026", 2026, "world_cup", "APARELHO", "Lihie Raz", "Israel", 12.383, None, None, 12.800, None],

    ["Cottbus World Cup 2026", 2026, "world_cup", "APARELHO", "Elisa Iorio", "Itália", None, 14.266, None, None, None],
    ["Cottbus World Cup 2026", 2026, "world_cup", "APARELHO", "Milana Kayumova", "Rússia", None, 14.133, 13.400, None, None],
    ["Cottbus World Cup 2026", 2026, "world_cup", "APARELHO", "Giulia Perotti", "Itália", None, 14.033, None, None, None],
    ["Cottbus World Cup 2026", 2026, "world_cup", "APARELHO", "Nakashima Karin", "Japão", None, 13.800, 10.566, None, None],
    ["Cottbus World Cup 2026", 2026, "world_cup", "APARELHO", "Lea Marie Quaas", "Alemanha", None, 13.033, 12.966, None, None],
    ["Cottbus World Cup 2026", 2026, "world_cup", "APARELHO", "Ulyana Kuzmenkova", "Belarus", None, 13.033, None, 12.333, None],
    ["Cottbus World Cup 2026", 2026, "world_cup", "APARELHO", "Kaylia Nemour", "Argélia", None, 12.900, 13.833, None, None],
    ["Cottbus World Cup 2026", 2026, "world_cup", "APARELHO", "Aiko Sugihara", "Japão", None, 12.133, 14.066, 13.133, None],

    ["Cottbus World Cup 2026", 2026, "world_cup", "APARELHO", "Bohdana Kovalova", "Ucrânia", None, None, 11.566, None, None],
    ["Cottbus World Cup 2026", 2026, "world_cup", "APARELHO", "Leila Vasilieva", "Rússia", None, None, 11.266, None, None],
    ["Cottbus World Cup 2026", 2026, "world_cup", "APARELHO", "Emma Puato", "Itália", None, None, None, 12.933, None],
    ["Cottbus World Cup 2026", 2026, "world_cup", "APARELHO", "Anastasiya Smantsar", "Belarus", None, None, None, 12.466, None],
    ["Cottbus World Cup 2026", 2026, "world_cup", "APARELHO", "Patricie Makovickova", "Tchéquia", None, None, None, 12.300, None],


    # BAKU WORLD CUP 2026

    ["Baku World Cup 2026", 2026, "world_cup", "APARELHO", "Anna Kalmykova", "Rússia", 13.716, 12.800, None, 13.366, None],
    ["Baku World Cup 2026", 2026, "world_cup", "APARELHO", "Tijana Korent", "Croácia", 13.133, None, None, None, None],
    ["Baku World Cup 2026", 2026, "world_cup", "APARELHO", "Miyata Shoko", "Japão", 13.016, None, 13.533, None, None],
    ["Baku World Cup 2026", 2026, "world_cup", "APARELHO", "Bohdana Kovalova", "Ucrânia", 12.933, 13.000, None, None, None],
    ["Baku World Cup 2026", 2026, "world_cup", "APARELHO", "Mila Prpic", "Croácia", 12.583, None, None, None, None],
    ["Baku World Cup 2026", 2026, "world_cup", "APARELHO", "Daniz Aliyeva", "Azerbaijão", 12.500, None, None, None, None],
    ["Baku World Cup 2026", 2026, "world_cup", "APARELHO", "Oksana Chusovitina", "Uzbequistão", 12.299, None, None, None, None],
    ["Baku World Cup 2026", 2026, "world_cup", "APARELHO", "Hillary Heron", "Panamá", 6.216, None, None, None, None],

    ["Baku World Cup 2026", 2026, "world_cup", "APARELHO", "Kaylia Nemour", "Argélia", None, 15.233, 13.800, None, None],
    ["Baku World Cup 2026", 2026, "world_cup", "APARELHO", "Leila Vasilieva", "Rússia", None, 14.033, None, None, None],
    ["Baku World Cup 2026", 2026, "world_cup", "APARELHO", "Sofia Shtykhetskaya", "Belarus", None, 13.800, None, None, None],
    ["Baku World Cup 2026", 2026, "world_cup", "APARELHO", "Ulyana Kuzmenkova", "Belarus", None, 12.666, None, None, None],
    ["Baku World Cup 2026", 2026, "world_cup", "APARELHO", "Magdalini Tsiori", "Grécia", None, 12.666, None, None, None],
    ["Baku World Cup 2026", 2026, "world_cup", "APARELHO", "Barbora Kundrát", "Eslováquia", None, 9.466, None, None, None],

    ["Baku World Cup 2026", 2026, "world_cup", "APARELHO", "Okamura Mana", "Japão", None, None, 14.133, 13.533, None],
    ["Baku World Cup 2026", 2026, "world_cup", "APARELHO", "Chiara Barzasi", "Itália", None, None, 13.233, None, None],
    ["Baku World Cup 2026", 2026, "world_cup", "APARELHO", "Anastasiia Lev", "Ucrânia", None, None, 12.833, None, None],
    ["Baku World Cup 2026", 2026, "world_cup", "APARELHO", "Angelica Finiguerra", "Itália", None, None, 12.800, None, None],
    ["Baku World Cup 2026", 2026, "world_cup", "APARELHO", "Philippa Busuttil", "Malta", None, None, 11.500, None, None],
    ["Baku World Cup 2026", 2026, "world_cup", "APARELHO", "Lucija Hribar", "Eslovênia", None, None, 10.800, None, None],

    ["Baku World Cup 2026", 2026, "world_cup", "APARELHO", "Sevgi Kayisoglu", "Turquia", None, None, None, 12.700, None],
    ["Baku World Cup 2026", 2026, "world_cup", "APARELHO", "Salina Bousmayo", "Marrocos", None, None, None, 12.333, None],
    ["Baku World Cup 2026", 2026, "world_cup", "APARELHO", "Anastasiya Smantsar", "Belarus", None, None, None, 12.266, None],
    ["Baku World Cup 2026", 2026, "world_cup", "APARELHO", "Kristina Hrudetska", "Ucrânia", None, None, None, 12.200, None],
    ["Baku World Cup 2026", 2026, "world_cup", "APARELHO", "Patricie Makovickova", "Tchéquia", None, None, None, 12.000, None],
    ["Baku World Cup 2026", 2026, "world_cup", "APARELHO", "Vita Prijanovic", "Eslovênia", None, None, None, 11.733, None],


    # CAIRO WORLD CUP 2026

    ["Cairo World Cup 2026", 2026, "world_cup", "APARELHO", "Laia Font", "Espanha", 13.533, None, 11.700, 12.900, None],
    ["Cairo World Cup 2026", 2026, "world_cup", "APARELHO", "Hillary Heron", "Panamá", 13.499, None, None, 12.933, None],
    ["Cairo World Cup 2026", 2026, "world_cup", "APARELHO", "Yu Linmin", "China", 13.349, None, None, None, None],
    ["Cairo World Cup 2026", 2026, "world_cup", "APARELHO", "Teja Belak", "Eslovênia", 13.333, None, None, None, None],
    ["Cairo World Cup 2026", 2026, "world_cup", "APARELHO", "Elizaveta Us", "Rússia", 13.299, None, 11.900, 12.600, None],
    ["Cairo World Cup 2026", 2026, "world_cup", "APARELHO", "Judy Abdalla", "Egito", 13.033, None, 11.866, None, None],
    ["Cairo World Cup 2026", 2026, "world_cup", "APARELHO", "Jiang Shuting", "China", 12.999, 13.700, None, None, None],
    ["Cairo World Cup 2026", 2026, "world_cup", "APARELHO", "Franchesca Santi", "Chile", 12.766, None, None, None, None],

    ["Cairo World Cup 2026", 2026, "world_cup", "APARELHO", "Kaylia Nemour", "Argélia", None, 14.033, 14.266, None, None],
    ["Cairo World Cup 2026", 2026, "world_cup", "APARELHO", "Lucija Hribar", "Eslovênia", None, 13.100, None, None, None],
    ["Cairo World Cup 2026", 2026, "world_cup", "APARELHO", "Ulyana Kuzmenkova", "Belarus", None, 12.933, None, None, None],
    ["Cairo World Cup 2026", 2026, "world_cup", "APARELHO", "Barbora Kundrát", "Eslováquia", None, 12.800, None, None, None],
    ["Cairo World Cup 2026", 2026, "world_cup", "APARELHO", "Sofia Mester", "Croácia", None, 12.733, None, None, None],
    ["Cairo World Cup 2026", 2026, "world_cup", "APARELHO", "Qiu Qiyuan", "China", None, 12.500, 12.833, None, None],
    ["Cairo World Cup 2026", 2026, "world_cup", "APARELHO", "Luiza Abel", "Brasil", None, 11.700, None, None, None],

    ["Cairo World Cup 2026", 2026, "world_cup", "APARELHO", "Ke Qinqin", "China", None, None, 13.166, 12.966, None],
    ["Cairo World Cup 2026", 2026, "world_cup", "APARELHO", "Anastasiya Smantsar", "Belarus", None, None, 11.766, None, None],
    ["Cairo World Cup 2026", 2026, "world_cup", "APARELHO", "Salina Bousmayo", "Marrocos", None, None, 9.433, 12.366, None],
    ["Cairo World Cup 2026", 2026, "world_cup", "APARELHO", "Thelma Adalsteinsdottir", "Islândia", None, None, None, 12.300, None],
    ["Cairo World Cup 2026", 2026, "world_cup", "APARELHO", "Bohdana Kovalova", "Ucrânia", None, None, None, 11.900, None],
    ["Cairo World Cup 2026", 2026, "world_cup", "APARELHO", "Ceren Biner", "Turquia", None, None, None, 11.866, None],


    # OSIJEK WORLD CUP 2026

    ["Osijek World Cup 2026", 2026, "world_cup", "APARELHO", "Karina Schönmaier", "Alemanha", 14.000, None, None, 12.533, None],
    ["Osijek World Cup 2026", 2026, "world_cup", "APARELHO", "Lia Monica Fontaine", "Canadá", 13.983, None, None, 13.100, None],
    ["Osijek World Cup 2026", 2026, "world_cup", "APARELHO", "Charlize Mörz", "Áustria", 13.716, None, None, 12.266, None],
    ["Osijek World Cup 2026", 2026, "world_cup", "APARELHO", "Hillary Heron", "Panamá", 13.449, None, None, None, None],
    ["Osijek World Cup 2026", 2026, "world_cup", "APARELHO", "Ruby Stacey", "Reino Unido", 13.316, None, 13.066, None, None],
    ["Osijek World Cup 2026", 2026, "world_cup", "APARELHO", "Athanasia Mesiri", "Grécia", 13.099, None, None, None, None],
    ["Osijek World Cup 2026", 2026, "world_cup", "APARELHO", "Benedetta Gava", "Itália", 12.766, None, None, 12.866, None],
    ["Osijek World Cup 2026", 2026, "world_cup", "APARELHO", "Teja Belak", "Eslovênia", 12.733, None, None, None, None],

    ["Osijek World Cup 2026", 2026, "world_cup", "APARELHO", "Yang Fanyuwei", "China", None, 15.266, None, None, None],
    ["Osijek World Cup 2026", 2026, "world_cup", "APARELHO", "Elena Colas", "França", None, 14.300, 13.433, 13.133, None],
    ["Osijek World Cup 2026", 2026, "world_cup", "APARELHO", "Vanesa Masova", "Tchéquia", None, 13.200, 13.033, None, None],
    ["Osijek World Cup 2026", 2026, "world_cup", "APARELHO", "Tian Zhuofan", "China", None, 13.066, 13.266, None, None],
    ["Osijek World Cup 2026", 2026, "world_cup", "APARELHO", "Nathalie Westlund", "Suécia", None, 12.933, None, None, None],
    ["Osijek World Cup 2026", 2026, "world_cup", "APARELHO", "Ekaterina Andreeva", "Rússia", None, 12.400, None, None, None],
    ["Osijek World Cup 2026", 2026, "world_cup", "APARELHO", "Zlata Osokina", "Rússia", None, 12.000, None, None, None],
    ["Osijek World Cup 2026", 2026, "world_cup", "APARELHO", "Céleste Mordenti", "Luxemburgo", None, 11.900, None, None, None],

    ["Osijek World Cup 2026", 2026, "world_cup", "APARELHO", "Manila Esposito", "Itália", None, None, 13.666, None, None],
    ["Osijek World Cup 2026", 2026, "world_cup", "APARELHO", "Zala Trtnik", "Eslovênia", None, None, 12.866, None, None],
    ["Osijek World Cup 2026", 2026, "world_cup", "APARELHO", "Ceren Biner", "Turquia", None, None, 11.600, None, None],
    ["Osijek World Cup 2026", 2026, "world_cup", "APARELHO", "Pauline Schäfer", "Alemanha", None, None, 11.166, None, None],

    ["Osijek World Cup 2026", 2026, "world_cup", "APARELHO", "Maïana Prat", "França", None, None, None, 13.100, None],
    ["Osijek World Cup 2026", 2026, "world_cup", "APARELHO", "July Marano", "Itália", None, None, None, 13.000, None],
    ["Osijek World Cup 2026", 2026, "world_cup", "APARELHO", "Charleen Pach", "Alemanha", None, None, None, 11.766, None],
]

df = pd.DataFrame(
    dados,
    columns=[
        "evento", "ano", "tipo_evento", "tipo_resultado",
        "ginasta", "nacionalidade", "VT", "UB", "BB", "FX", "AA"
    ]
)


# SOMA DOS 4 APARELHOS


df["tem_4_aparelhos"] = df[["VT", "UB", "BB", "FX"]].notna().all(axis=1)

df["soma_4_aparelhos"] = np.where(
    df["tem_4_aparelhos"],
    df[["VT", "UB", "BB", "FX"]].sum(axis=1),
    np.nan
)

df["nota_modelo"] = df["AA"].fillna(df["soma_4_aparelhos"])

# manter ginastas com pelo menos uma nota AA ou soma dos 4 aparelhos

ginastas_validas = df.loc[df["nota_modelo"].notna(), "ginasta"].unique()
df_modelagem = df[df["ginasta"].isin(ginastas_validas)].copy()

# PESOS

peso_evento = {
    "mundial": 1.00,
    "continental": 0.75,
    "internacional": 0.60,
    "world_cup": 0.45
}

df_modelagem["peso_evento"] = df_modelagem["tipo_evento"].map(peso_evento)

df_modelagem["peso_recencia"] = np.where(df_modelagem["ano"] == 2026, 1.20, 1.00)

df_modelagem["peso_tipo"] = np.where(
    df_modelagem["AA"].notna(),
    1.00,
    0.70
)

df_modelagem["peso_final"] = (
    df_modelagem["peso_evento"] *
    df_modelagem["peso_recencia"] *
    df_modelagem["peso_tipo"]
)
```

# Modelagem

```python
# FEATURES POR GINASTA

base = (
    df_modelagem[df_modelagem["nota_modelo"].notna()]
    .groupby(["ginasta", "nacionalidade"])
    .apply(lambda x: pd.Series({
        "media_ponderada": np.average(x["nota_modelo"], weights=x["peso_final"]),
        "melhor_nota": x["nota_modelo"].max(),
        "pior_nota": x["nota_modelo"].min(),
        "desvio": x["nota_modelo"].std(),
        "qtd_resultados": len(x),
        "qtd_AA_real": x["AA"].notna().sum(),
        "qtd_soma_4_aparelhos": x["soma_4_aparelhos"].notna().sum(),
        "ultima_nota": x.sort_values("ano")["nota_modelo"].iloc[-1]
    }))
    .reset_index()
)

base["desvio"] = base["desvio"].fillna(0.60)

# FORMA RECENTE POR APARELHO

ap_long = df_modelagem.melt(
    id_vars=["evento", "ano", "tipo_evento", "ginasta", "nacionalidade"],
    value_vars=["VT", "UB", "BB", "FX"],
    var_name="aparelho",
    value_name="nota_aparelho"
).dropna(subset=["nota_aparelho"])

forma_aparelho = (
    ap_long
    .groupby(["ginasta", "nacionalidade"])
    .agg(
        media_aparelhos=("nota_aparelho", "mean"),
        melhor_aparelho=("nota_aparelho", "max"),
        qtd_notas_aparelho=("nota_aparelho", "count")
    )
    .reset_index()
)

base = base.merge(
    forma_aparelho,
    on=["ginasta", "nacionalidade"],
    how="left"
)

base[["media_aparelhos", "melhor_aparelho", "qtd_notas_aparelho"]] = (
    base[["media_aparelhos", "melhor_aparelho", "qtd_notas_aparelho"]]
    .fillna(0)
)

# SCORE PREDITIVO

base["bonus_forma_aparelhos"] = np.where(
    base["media_aparelhos"] > 0,
    (base["media_aparelhos"] - 13.20) * 0.20,
    0
)

base["bonus_evidencia_AA"] = base["qtd_AA_real"] * 0.05

base["penalizacao_instabilidade"] = base["desvio"] * 0.20

base["penalizacao_pouca_base"] = 0.30 / np.sqrt(base["qtd_resultados"])

base["score_projetado"] = (
    0.65 * base["media_ponderada"]
    + 0.20 * base["melhor_nota"]
    + 0.10 * base["ultima_nota"]
    + base["bonus_forma_aparelhos"]
    + base["bonus_evidencia_AA"]
    - base["penalizacao_instabilidade"]
    - base["penalizacao_pouca_base"]
)


# PROBABILIDADE VIA SOFTMAX

temperatura = 0.55

base["prob_vitoria"] = np.exp(
    (base["score_projetado"] - base["score_projetado"].max()) / temperatura
)

base["prob_vitoria"] = base["prob_vitoria"] / base["prob_vitoria"].sum()

ranking = base.sort_values("prob_vitoria", ascending=False).copy()
ranking["prob_vitoria_pct"] = (ranking["prob_vitoria"] * 100).round(2)


# RESULTADOS

print("BASE COMPLETA")
display(df)

print("BASE USADA NO MODELO")
display(df_modelagem)

print("TOP 3 - PREDIÇÃO ROTTERDAM 2026")
display(
    ranking[
        [
            "ginasta",
            "nacionalidade",
            "media_ponderada",
            "melhor_nota",
            "ultima_nota",
            "media_aparelhos",
            "qtd_AA_real",
            "qtd_soma_4_aparelhos",
            "score_projetado",
            "prob_vitoria_pct"
        ]
    ].head(3)
)

# EXPORTAR EXCEL

with pd.ExcelWriter("modelo_ginastica_rotterdam_2026_corrigido.xlsx", engine="openpyxl") as writer:
    df.to_excel(writer, sheet_name="base_total", index=False)
    df_modelagem.to_excel(writer, sheet_name="base_modelagem", index=False)
    ranking.to_excel(writer, sheet_name="ranking_predicao", index=False)

print("Arquivo salvo: modelo_ginastica_rotterdam_2026_corrigido.xlsx")
```
